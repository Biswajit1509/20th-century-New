```python
# Cell A — open the page and confirm title (uses the webdriver-manager driver)
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time, os

options = Options()
options.add_argument("--headless")   # keep headless for background run
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# exact page used in the exercise (change only if instructor specified another URL)
page_url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century"
driver.get(page_url)
time.sleep(2)   # small wait to ensure JS/DOM settled

print("Loaded page title:", driver.title)
# show working directory
print("Working dir:", os.getcwd())
```

```python
# Cell B — save raw HTML and visible text (use BeautifulSoup for text extraction)
from bs4 import BeautifulSoup

page_source = driver.page_source

# 1) Save raw HTML (backup)
with open("20th_century_page_raw.html", "w", encoding="utf-8") as f:
    f.write(page_source)

# 2) Parse and save visible text (chapter uses soup.get_text)
soup = BeautifulSoup(page_source, "html.parser")
text_content = soup.get_text(separator="\n")

with open("20th_century_content.txt", "w", encoding="utf-8") as f:
    f.write(text_content)

print("Saved files: 20th_century_page_raw.html and 20th_century_content.txt")
```
