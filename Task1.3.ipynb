```python
import pandas as pd

# Create test DataFrame with at least one text and one numeric column
data = pd.DataFrame({
    "Name": ['Aragorn', 'Boromir', 'Frodo', 'Galadriel', 'Legolas'],
    "Race": ["human", "human", "hobbit", "elf", "elf"],
    "Age": [88, 41, 51, 8382, 293],        # numeric
    "Score": [9.5, 7.2, 8.1, 9.9, 9.1]     # numeric
})

# Display the DataFrame
data
```

```python
# Example: divide Score by 2 and save it to a new column
data['score_div_2'] = data['Score'].apply(lambda x: x / 2)

# Or using vectorized operation (faster)
# data['score_div_2'] = data['Score'] / 2

data.head()
```

```python
# Create a boolean flag if Race == 'elf'
data['is_elf'] = data['Race'].apply(lambda x: True if x == 'elf' else False)

data[['Name', 'Race', 'is_elf']].head()
```

```python
# Create a labelled flag column from Age
data['age_label'] = data['Age'].apply(
    lambda x: 'elven old' if x > 500 else ('middle-aged' if x >= 100 else 'young')
)

data[['Name', 'Age', 'age_label']].head()
```
